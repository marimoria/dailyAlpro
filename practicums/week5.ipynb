# 05A. Rataan Bilangan pada Posisi Tertentu
### Description
Buat program membaca n bilangan bulat dan menghitung rata-rata dari bilangan-bilangan yang berada di posisi kelipatan k. Jika n = 0 atau tidak ada data pada posisi tersebut, maka rata-rata bernilai 0.

### Input Format
Baris pertama berisi 2 bilangan bulat n (0 ≤ n ≤ 100000) dan k (0 ≤ k ≤ 100000), dimana n adalah banyaknya bilangan, dan k adalah posisi kelipatan ke-k yang ingin dihitung rata-ratanya. Bilangan pertama menempati posisi ke-1, bilangan kedua posisi ke-2, dan seterusnya. Baris berikutnya berisi n bilangan bulat.

### Output Format
Sebuah bilangan double yang menunjukkan nilai rata-rata dari data pada posisi kelipatan k.

### Sample Input
```
7 2
10 20 30 40 50 60 70
```
### Sample Output
```
40.00
```

### Thought Process
Kita memproses data secara *on-the-fly* menggunakan loop `for (int i = 1; i <= n; i++)`. Jika iterasi `i % k == 0`, kita tambahkan data tersebut ke `sum` dan naikkan `count`. Di akhir, hati-hati terhadap *division by zero* jika `n == 0`, `k == 0`, atau `count == 0`.

````python
%%writefile practicumWeek5a.c
#include <stdio.h>

int main() {
  int n, k;
  int nilai;
  int sum = 0;
  int count = 0;
  double average;
  
  scanf("%d %d", &n, &k);
  
  if (n == 0 || k == 0) {
    average = 0;
    printf("%.2lf\n", average);
    return 0;
  }
  
  for (int i = 1; i <= n; i++) {
    scanf("%d", &nilai);
    if (i % k == 0) {
      sum += nilai;
      count++;
    }
  }
  
  if (count == 0) {
    average = 0;
  } else {
    average = (sum*1.0) / count;
  }
  
  printf("%.2lf\n", average);
  return 0;
}
````

# 05B. Bilangan Palindrome
### Description
Palindrome adalah kata, rangkaian kata, atau bilangan yang terbaca sama baik dari depan maupun dari belakang. Dengan kata lain, palindrome adalah suatu kata, frasa, angka, maupun susunan lainnya yang apabila dibaca dari depan maupun belakang akan memiliki bunyi yang sama. Sebagai contoh, bilangan 232 adalah palindrome, sedangkan 234 adalah bukan palindrome.
Buatlah program untuk membaca n bilangan bulat, dan periksalah apakah setiap bilangan tersebut merupakan palindrome atau bukan.

### Input Format
Baris pertama adalah n (1<n<5000) yang menunjukkan banyaknya bilangan yang akan diperiksa. Baris-baris berikutnya sebanyak n bilangan bulat positif.

### Output Format
Sebanyak n baris hasil pemeriksaan bilangan. Jika palindrome, tuliskan angka 1, sebaliknya tuliskan angka 0.

### Sample Input
```
5
2334332
555
1234
50
123
```
### Sample Output
```
1
1
0
0
0
```

### Thought Process
Metode paling efisien adalah membalikkan **setengah bagian** dari angka, lalu membandingkannya dengan setengah sisanya.
1. Ambil digit terakhir: `lastDigit = x % 10`
2. Masukkan ke hasil terbalik: `final = final * 10 + lastDigit`
3. Buang digit dari angka asli: `x = x / 10`
Berhenti memotong saat `final >= x`.
Untuk angka genap (1221), cek `final == x`. Untuk ganjil (12321), cek `final / 10 == x` (membuang digit tengah).

````python
import matplotlib.pyplot as plt

def visualize_palindrome(num):
    fig, ax = plt.subplots(figsize=(6, 2))
    s = str(num)
    for i, char in enumerate(s):
        ax.text(i, 0.5, char, fontsize=20, ha='center', 
                bbox=dict(facecolor='lightblue' if i < len(s)/2 else 'lightgreen', alpha=0.5))
    ax.set_xlim(-1, len(s))
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title(f"Visualisasi Pemotongan Setengah: {num}")
    plt.show()

visualize_palindrome(2334332)
````

````python
%%writefile practicumWeek5b.c
# include <stdio.h>

int main() {
  int n;
  scanf("%d", &n);
  
  for (int i = 0; i < n; i++) {
    int x;
    int final = 0;
    int lastDigit;
    scanf("%d", &x);
    
    // Eliminasi awal untuk jebakan (negatif dan akhiran 0)
    if (x < 0 || (x % 10 == 0 && x != 0)) {
      printf("0\n");
    } else {
        while (final < x) {
          lastDigit = x % 10;
          x = x / 10;
          final = final*10 + lastDigit;
        }
        
        if (final == x) {
          printf("1\n");
        } else if (final / 10 == x) {
          printf("1\n");
        } else {
          printf("0\n");
        }
    }
  }
  return 0;
}
````
