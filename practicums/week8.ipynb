## Question 1
### Description
Buat program membaca N bilangan bulat yang masing-masing memiliki kisaran nilai [-100,100], dan hitung frekuensi setiap bilangan. Tampilan frekuensi setiap bilangan terurut dari kecil ke besar.

### Constraints
Tidak boleh ada proses sorting (mengurutkan) data.

$1 ≤ N ≤ 100.000$

### Input
Input data diberikan dengan format berikut:
```
N
A1 A2 … AN
```

### Output
Frekuensi setiap data dengan format:
```
Ai ada Ni
```

### Sample Input
```
5 
-2 14 -2 5 -2 
```

### Sample Output
```
-2 ada 3
5 ada 1
14 ada 1
```

````python
%%writefile week8_A.c
# include <stdio.h>

int main() {
  int jumlahAngka;
  int angka = 0;

  /*
    angka mines di baris 0 angka positif di baris 1
    -1 di daftarAngka[0][1], -2 di [0][2], -100 di [0][100]
    1 di daftarAngka[1][1], 2 di [1][2], 100 di [1][100]
    
    angka mines ada -1 s.d -100 = 100 index -> 101 elemen
    angka positif ada 0 s.d 100 = 100 index -> 101 elemen
  */ 
  int daftarAngka[2][101] = {0};

  scanf("%d", &jumlahAngka);

  for (int i = 0; i < jumlahAngka; i++) {
    scanf("%d", &angka);

    if (angka < 0) {
      daftarAngka[0][(angka * (-1))]++;
    } else {
      daftarAngka[1][angka]++;
    }
  }

  for (int i = 0; i < 2; i++) {
    if (i == 0) {
      for (int j = 100; j >= 1; j--) {
        if (daftarAngka[i][j] != 0) {
          printf("%d ada %d\n", (j * (-1)), daftarAngka[i][j]);
        }
      }
    } else {
      for (int j = 0; j < 101; j++) {
        if (daftarAngka[i][j] != 0) {
          printf("%d ada %d\n", j, daftarAngka[i][j]);
        }
      }
    }
  }

  return 0;
}
````

## Question 2
### Description
Buat program membaca matrik berukuran $M×N$, dan menampilkan matrik setelah dilakukan flip secara vertikal.

### Constraints
$1≤M$, $N≤1000$

### Input
Input data dengan format:
```
M N
A11 A12 … A1N
A21 A22 … A2N
…
AM1 AM2 … AMN
```

### Output
```
Matrik setelah dilakukan flip vertikal.
```

### Sample Input
```
4 5
1 0 8 6 1 
0 0 0 0 2 
0 0 0 3 3 
4 5 1 2 4
```

### Sample Output
```
4 5 1 2 4 
0 0 0 3 3 
0 0 0 0 2 
1 0 8 6 1
```

````python
%%writefile week8_B.c
# include <stdio.h>

int matrix[1000][1000] = {0};
int main() {
  int baris, kolom;

  scanf("%d %d", &baris, &kolom);

  for (int i = 0; i < baris; i++) {
    for (int j = 0; j < kolom; j++) {
      scanf("%d", &matrix[i][j]);
    }
  }

  for (int i = (baris - 1); i >= 0; i--) {
    for (int j = 0; j < kolom; j++) {
      printf("%d", matrix[i][j]);

      if (j == (kolom - 1)) {
        printf("\n");
      } else {
        printf(" ");
      }
    }
  }

  return 0;
}
````
