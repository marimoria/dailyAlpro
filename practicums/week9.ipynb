# 09A. Angka Keramat 47
## Description
47 merupakan angka keramat bagi Mukidi. Oleh karena itu, setiap menemukan sekumpulan angka, Mukidi selalu mencari angka 47 atau mendapatkan sekumpulan bilangan yang bersebelahan (continuous subsequences) yang jumlahnya adalah 47.

Saat ini, Mukidi melihat ada sederetan angka, misalnya {24, 17, 23, 24, 5, 30, 47}. Maka Mukidi mulai mencarinya, dan ditemukan ada 2, yaitu {23, 24} dan {47}. Deretan {17, 30} tidak termasuk karena mereka bukan continuous subsequences.

Bantulah Mukidi dengan membuat program untuk menghitung banyaknya kemungkinan angka keramat yang ada di dalam N deretan angka bilangan bulat positif.

## Constraints
$$1≤N≤100000$$
$$1≤A_i≤25$$

## Input
Input data diberikan dengan format berikut:
```
N
A1 A2 … AN
```

## Output
Sebuah bilangan bulat yang menunjukkan banyaknya kemungkinan angka keramat 47 sesuai dengan deskripsi soal.

### Sample Input 1
```
7
24 17 23 24 5 30 47
```
### Sample Output 1
```
2
```

### Sample Input 2
```
7
2 47 10047 47 1047 47 47
```

### Sample Output 2
```
4
```

````python
%%writefile week9_A.c
#include <stdio.h>

int daftarAngka[100000];

int main() {
  int n;
  long long count = 0;
  long long sum = 0;
  int start = 0;
  int prefixZeroes = 0;

  scanf("%d", &n);

  for (int i = 0; i < n; i++) {
    scanf("%d", &daftarAngka[i]);
  }

  for (int end = 0; end < n; end++) {
    sum += daftarAngka[end];

    while (sum > 47 && start <= end) {
      sum -= daftarAngka[start];
      start++;
      prefixZeroes = 0;
    }

    while (sum == 47 && start <= end && daftarAngka[start] == 0) {
      prefixZeroes++;
      sum -= daftarAngka[start];
      start++;
    }

    if (sum == 47) {
      count += 1 + prefixZeroes;
    }
  }

  printf("%lld\n", count);
  return 0;
}
````

# 09B. Prefix Sum 1D

## Description
Prefix sum dari array `arr` adalah array lain (`prefixSum`) berukuran sama, dimana:

  prefixSum[i] = arr[0] + arr[1] + ... + arr[i]

Diberikan array `prefixSum` berukuran N, tentukan nilai `arr` pada posisi-posisi tertentu.

## Constraints
- 1 <= N <= 100,000

## Input
  N
  A1 A2 ... AN
  P1 P2 ... -9

- Ai adalah prefix sum pada indeks ke-i
- Query diakhiri dengan -9

## Output
Nilai arr[Pi] untuk setiap query Pi (posisi berbasis 1), dipisahkan spasi.

## Key Formula
- arr[0] = prefixSum[0]
- arr[i] = prefixSum[i] - prefixSum[i-1] untuk i > 0

## Sample
Input:
```
  10
  3 10 26 30 46 55 64 65 80 102
  1 5 3 2 7 -9
```

Output:
```
  3 16 16 7 9
```

````python
%%writefile week9_B.c
# include <stdio.h>

int main() {
  int n;
  int arrayAsli[100005] = {0};
  int PS[100005] = {0};

  scanf("%d", &n);

  for (int i = 1; i <= n; i++) {
    scanf("%d", &PS[i]);
  }

  int indexAsli;
  scanf("%d", &indexAsli);

  while (indexAsli != -9) {
    if (indexAsli == 1) {
      printf("%d ", PS[1]);
    } else {
      printf("%d ", PS[indexAsli] - PS[indexAsli-1]);
    }
    
    scanf("%d", &indexAsli);
  }

  return 0;
}
````

# 09C. Prefix Sum 2D

Time limit: 250 ms
Memory limit: 10 MB

## Description
Diketahui prefix sum 2 dimensi berukuran M x N seperti yang dijelaskan di slide kuliah.
Tuliskan nilai asli dari array yang ada pada posisi baris ke-B dan kolom ke-K.
Data array paling atas kiri dianggap sebagai posisi baris ke-1 kolom ke-1.
Pertanyaan posisi data awal ini diakhiri dengan nilai sentinel -9.

## Constraints
- 1 <= M, N <= 1000

## Input
- Baris pertama adalah nilai M dan N, kemudian diikuti dengan input data
  dalam bentuk matrik berukuran M x N.
- Baris-baris berikutnya adalah pertanyaan posisi data asli pada baris ke-B
  dan kolom ke-K, yang diakhiri dengan -9.

## Output
Beberapa bilangan sesuai dengan pertanyaan.

## Sample
Input:
```
  3 3
  10 30 60
  15 45 95
  17 51 107
  1 1
  2 1
  1 3
  2 2
  -9
```

Output:
```
  10 5 30 6
```

````python
%%writefile week9_C.c
#include <stdio.h>

int PS[1005][1005] = {0};

int main() {
  int m, n;
  int b, k;

  if (scanf("%d %d", &m, &n) != 2) return 0;

  for (int i = 1; i <= m; i++) {
    for (int j = 1; j <= n; j++) {
      scanf("%d", &PS[i][j]);
    }
  }

  // Reads b, checks if it's -9, then reads k
  while (scanf("%d", &b) == 1 && b != -9 && scanf("%d", &k) == 1) {
    if (b == 1 && k == 1) {
      printf("%d ", PS[1][1]);
    } else if (b == 1) {
      printf("%d ", PS[b][k] - PS[b][k-1]);
    } else if (k == 1) {
      printf("%d ", PS[b][k] - PS[b-1][k]);
    } else {
      printf("%d ", PS[b][k] - PS[b-1][k] - PS[b][k-1] + PS[b-1][k-1]);
    }
  }

  return 0;
}
````
