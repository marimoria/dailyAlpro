# 01B. Baca Tulis Angka
### Description
Program tidak hanya dapat mencetak namun dapat membaca. Pada soal ini, Anda diminta untuk membuat program yang dapat membaca sebuah bilangan bulat dan mencetaknya kembali.

Salah satu fungsi yang dapat digunakan untuk membaca nilai masukan adalah scanf. Anda dapat kunjungi dokumentasi bahasa C untuk melihat kapabilitas fungsi scanf beserta dengan contoh-contoh pemakaiannya.

### Input Format
Sebuah bilangan bulat dengan nilai yang berada pada rentang 1 hingga 10 milyar. 

### Output Format
Sebuah bilangan bulat dengan nilai yang berada pada rentang 1 hingga 10 milyar yang dimasukkan ke program. Akhiri keluaran dengan karakater newline ('\n').

### Sample Input 1
```
2023
```
### Sample Output 1
```
2023
```

### Sample Input 2
```
9876543210
```
### Sample Output 2
```
9876543210
```

Petunjuk
Perhatikan ukuran masukan, dalam hal ini nilai dari bilangan bulat yang Anda masukkan pada program.

````python
%%writefile practicumWeek1b.c
# include <stdio.h>

int main() {
  unsigned long long x;

  scanf("%llu", &x);
  printf("%llu\n", x);
  
  return 0;
}
````

````python
!gcc practicumWeek1b.c -o practicumWeek1b && echo 2023 | practicumWeek1b.exe
````

# 01C. Pajak Penjualan
### Description
Setiap pelanggan warung makanan dikenakan pajak 15% dari nilai pembelian. Buatlah program untuk mendapatkan nilai pembelian yang sebenarnya jika diketahui pelanggan membayar belanja makanan di kasir dengan nilai uang tertentu.

### Input Format
Sebuah bilangan double, C (0<C<2M), yang menunjukkan nilai yang dibayarkan oleh pembeli (rupiah).

### Output Format
Sebuah bilangan double yang menunjukkan nilai pembelian sebenarnya (2 digit di belakang tanda desimal) yang diakhiri dengan newline.

### Sample Input
```
1150
```
### Sample Output
```
1000.00
```

````python
%%writefile practicumWeek1c.c
# include <stdio.h>

int main() {
  double hargaFinal, hargaAsli;
  int persenPajak;
  persenPajak = 15;

  scanf("%lf", &hargaFinal);

  /*
    Karena hargaFinal = hargaAsli + (persenPajak/100 * hargaAsli) maka
           hargaFinal = (100*hargaAsli + persenPajak*hargaAsli) / 100
           hargaFinal = [(100 + persenPajak) * hargaAsli]/100
           hargaAsli = (100 * hargaFinal) / (100 + persenPajak)
  */
  hargaAsli = ((100 * hargaFinal) / (100 + persenPajak));

  printf("%.2lf\n", hargaAsli);
  
  return 0;
}
````

````python
!echo 1150 | gcc practicumWeek1c.c -o practicumWeek1c && echo 1150 | practicumWeek1c.exe
````
